```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

# Loading the data. 
The parameter return_X_y=True makes the function return the data and the target as separate variables.

```python
X, y = datasets.load_diabetes(return_X_y=True)
```

# Preparing the data for training.
We are making a column matrix of the feature data for 2 examples. The first one is plotting diabetes progression after BMI and the second one is plotting diabetes progression after Low density lipoprotein cholesterol.

```python
X_BMI = X[:, np.newaxis, 2]
X_S2LDL = X[:, np.newaxis, 5]
```

# Choosing training method and training the model.
Creating the model for the first example.

```python
X_BMI_train, X_BMI_test, y_train, y_test = model_selection.train_test_split(X_BMI, y, test_size=0.33) # 33% test data
model = linear_model.LinearRegression() # create linear regression object
model.fit(X_BMI_train, y_train) # train the model using the training sets
y_pred = model.predict(X_BMI_test) # make predictions using the testing set
```

Creating the plot for the first example.

```python
plt.scatter(X_BMI_test, y_test,  color='black')
plt.plot(X_BMI_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

Creating the model for the second example.

```python
X_S2LDL_train, X_S2LDL_test, y_train_S2LDL, y_test_S2LDL = model_selection.train_test_split(X_S2LDL, y, test_size=0.33) # 33% test data
model_S2LDL = linear_model.LinearRegression() # create linear regression object
model_S2LDL.fit(X_S2LDL_train, y_train_S2LDL) # train the model using the training sets
y_pred_S2LDL = model_S2LDL.predict(X_S2LDL_test) # make predictions using the testing set
```

Creating the plot for the second example.

```python
plt.scatter(X_S2LDL_test, y_test_S2LDL,  color='black')
plt.plot(X_S2LDL_test, y_pred_S2LDL, color='blue', linewidth=3)
plt.xlabel('Scaled S2LDL')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against S2LDL')
plt.show()
```

The blue lines are a collection of points that are the result of the linear regression model.
